```python
def show_all_columns(df):
    with pd.option_context('display.max_columns', None):
        display(df)
def show_all_rows(df):
    with pd.option_context('display.max_rows', None):
        display(df)
```

```python
dfmajorcrimes = pd.read_csv('raw_data/majorcrimes.csv')
dfhomicides = pd.read_csv('raw_data/homicidies.csv')
```

```python
# Define the target columns including the ones that need to be added
columns = ['EVENT_UNIQUE_ID','DATASET','OFFENCE', 'MCI_CATEGORY','OCC_HOUR','OCC_DAY','OCC_MONTH','OCC_YEAR','OCC_DOW','OCC_DOY','LAT_WGS84','LONG_WGS84']

# Add missing columns with default values
for df in [dfhomicides, dfmajorcrimes]:
    for column in columns:
        if column not in df.columns:
            df[column] = None

# Select columns in the specified order for all DataFrames
dfhomicides_s = dfhomicides[columns]
dfmajorcrimes_s = dfmajorcrimes[columns]

# Set the 'DATASET' column for each DataFrame using .loc to avoid SettingWithCopyWarning
dfhomicides_s.loc[:, 'DATASET'] = 'HOMICIDES'
dfmajorcrimes_s.loc[:, 'DATASET'] = 'MAJOR_CRIMES'

# Concatenate the DataFrames into a single DataFrame
dfmatrix = pd.concat([dfhomicides_s, dfmajorcrimes_s], ignore_index=True)

# Display the resulting DataFrame
dfmatrix
```

```python
dfmatrix.loc[dfmatrix['DATASET'] == 'HOMICIDES', 'OFFENCE'] = 'Homicide'
dfmatrix.loc[dfmatrix['DATASET'] == 'HOMICIDES', 'MCI_CATEGORY'] = 'Homicide'
dfmatrix
```

# Data Cleaning and Transformation

```python
# checking for latitude and longitude errors

# Toronto boundaries
north_boundary = 43.8554
south_boundary = 43.5810
east_boundary = -79.1161
west_boundary = -79.6393


# Check if latitude and longitude are within the boundaries
within_boundaries = (dfmatrix['LAT_WGS84'] <= north_boundary) & (dfmatrix['LAT_WGS84'] >= south_boundary) & \
                    (dfmatrix['LONG_WGS84'] >= west_boundary) & (dfmatrix['LONG_WGS84'] <= east_boundary)

# Filter the DataFrame
dfmatrix = dfmatrix.loc[within_boundaries].reset_index(drop=True)
dfmatrix
```

### Fixing OCC_DAY

```python
#  Converting OCC_DAY from float to int
dfmatrix['OCC_DAY'] = pd.to_numeric(dfmatrix['OCC_DAY'], errors='coerce').fillna(0).astype(int)
```

```python
# Apply np.floor and convert to int, while handling NaN values
dfmatrix['OCC_DAY'] = dfmatrix['OCC_DAY'].apply(lambda x: np.floor(x) if pd.notnull(x) else x).astype('Int64')
```

```python
# Filter the DataFrame for non-zero OCC_DAY values
df_occ_day_non_zero = dfmatrix[dfmatrix['OCC_DAY'] != 0]

# Calculate the value counts and normalize them
occ_day_proportions = df_occ_day_non_zero['OCC_DAY'].value_counts(normalize=True).reset_index()
occ_day_proportions.columns = ['OCC_DAY', 'proportion']

# Total number of rows where OCC_DAY is 0
total_zeros = (dfmatrix['OCC_DAY'] == 0).sum()

# Track the index of the current proportion
current_proportion_index = 0

# Iterate over the rows where OCC_DAY is 0 and update OCC_DAY based on proportions
for index, row in dfmatrix[dfmatrix['OCC_DAY'] == 0].iterrows():
    # Get the current proportion
    current_proportion = occ_day_proportions.at[current_proportion_index, 'OCC_DAY']
    
    # Update OCC_DAY for the current row
    dfmatrix.at[index, 'OCC_DAY'] = current_proportion
    
    # Update the current proportion index
    current_proportion_index += 1
    
    # If we reach the end of the proportions, reset the index
    if current_proportion_index >= len(occ_day_proportions):
        current_proportion_index = 0
```

```python
# Dropping the zero rows
# Total number of rows
total_rows = len(dfmatrix)

# Number of rows where OCC_DAY is 0
occ_day_zeros = (dfmatrix['OCC_DAY'] == 0).sum()

# Proportion of rows where OCC_DAY is 0
proportion_occ_day_zeros = occ_day_zeros / total_rows

print(f"Total rows: {total_rows}")
print(f"Rows with OCC_DAY = 0: {occ_day_zeros}")
print(f"Proportion of rows with OCC_DAY = 0: {proportion_occ_day_zeros:.2%}")
```

### OCC_DAY E DOY para Sin e COS


```python
dfmatrix['OCC_DAY_SIN'] = np.sin(2 * np.pi * dfmatrix['OCC_DAY'] / 24)
dfmatrix['OCC_DAY_COS'] = np.cos(2 * np.pi * dfmatrix['OCC_DAY'] / 24)

dfmatrix['OCC_DOY_SIN'] = np.sin(2 * np.pi * dfmatrix['OCC_DOY'] / 24)
dfmatrix['OCC_DOY_COS'] = np.cos(2 * np.pi * dfmatrix['OCC_DOY'] / 24)
```

```python
months = {'January':1,
          'February': 2,
          'March': 3,
          'April':4,
          'May': 5,
          'June': 6,
          'July': 7,
          'August': 8,
          'September': 9,
          'October': 10,
          'November': 11,
          'December': 12}
```

```python
dfmatrix['OCC_MONTH_NUM'] = dfmatrix.OCC_MONTH.apply(lambda x: months[x])
```

```python
# Apply sine and cosine transformations
dfmatrix['MONTH_SIN'] = np.sin(2 * np.pi * dfmatrix['OCC_MONTH_NUM'] / 12)
dfmatrix['MONTH_COS'] = np.cos(2 * np.pi * dfmatrix['OCC_MONTH_NUM'] / 12)
dfmatrix['HOUR_SIN'] = np.sin(2 * np.pi * dfmatrix['OCC_HOUR'] / 24)
dfmatrix['HOUR_COS'] = np.cos(2 * np.pi * dfmatrix['OCC_HOUR'] / 24)
```

```python
days_of_week = {'Friday': 6,  
                'Saturday': 7,
                'Sunday': 1,  
                'Thursday': 5,
                'Wednesday': 4, 
                'Monday': 2,
                'Tuesday': 3}

dfmatrix['OCC_DOW_NUM'] = dfmatrix.OCC_DOW.apply(lambda x: days_of_week[x.strip()])

dfmatrix['OCC_DOW_SIN'] = np.sin(2 * np.pi * dfmatrix['OCC_DOW_NUM'] / 24)
dfmatrix['OCC_DOW_COS'] = np.cos(2 * np.pi * dfmatrix['OCC_DOW_NUM'] / 24)
```

```python
# Função para criar o modelo
def build_model(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers', 1, 6)):
        model.add(LSTM(units=hp.Int(f'units_{i}', min_value=16, max_value=512, step=32), return_sequences=True if i < hp.get('num_layers') - 1 else False))
        model.add(Dropout(hp.Float(f'dropout_{i}', min_value=0.0, max_value=0.5, step=0.1)))
        model.add(BatchNormalization())
    model.add(Dense(units=hp.Int('dense_units', min_value=16, max_value=256, step=32), activation='relu'))
    model.add(Dropout(hp.Float('dense_dropout', min_value=0.0, max_value=0.5, step=0.1)))
    model.add(Dense(len(y.columns), activation='softmax'))
    
    model.compile(optimizer=Adam(hp.Float('learning_rate', min_value=1e-5, max_value=1e-2, sampling='LOG')),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Criar o tuner
tuner = kt.RandomSearch(build_model,
                        objective='val_accuracy',
                        max_trials=100,  # Aumente o número de testes para uma busca mais abrangente
                        executions_per_trial=1,
                        directory='my_dir',
                        project_name='LSTM_tuning')

# Parada antecipada
stop_early = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Executar a busca
tuner.search(X_train, y_train, epochs=50, validation_split=0.2, callbacks=[stop_early])

# Obter os melhores hiperparâmetros
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"""
The hyperparameter search is complete. The optimal number of layers is {best_hps.get('num_layers')}.
The optimal number of units in the first LSTM layer is {best_hps.get('units_0')},
in the second LSTM layer is {best_hps.get('units_1') if best_hps.get('num_layers') > 1 else 'N/A'},
in the third LSTM layer is {best_hps.get('units_2') if best_hps.get('num_layers') > 2 else 'N/A'},
in the fourth LSTM layer is {best_hps.get('units_3') if best_hps.get('num_layers') > 3 else 'N/A'},
in the fifth LSTM layer is {best_hps.get('units_4') if best_hps.get('num_layers') > 4 else 'N/A'},
in the sixth LSTM layer is {best_hps.get('units_5') if best_hps.get('num_layers') > 5 else 'N/A'},
the optimal number of units in the dense layer is {best_hps.get('dense_units')},
the optimal dropout rate in the first LSTM layer is {best_hps.get('dropout_0')},
the optimal dropout rate in the second LSTM layer is {best_hps.get('dropout_1') if best_hps.get('num_layers') > 1 else 'N/A'},
the optimal dropout rate in the third LSTM layer is {best_hps.get('dropout_2') if best_hps.get('num_layers') > 2 else 'N/A'},
the optimal dropout rate in the fourth LSTM layer is {best_hps.get('dropout_3') if best_hps.get('num_layers') > 3 else 'N/A'},
the optimal dropout rate in the fifth LSTM layer is {best_hps.get('dropout_4') if best_hps.get('num_layers') > 4 else 'N/A'},
the optimal dropout rate in the sixth LSTM layer is {best_hps.get('dropout_5') if best_hps.get('num_layers') > 5 else 'N/A'},
the optimal dropout rate in the dense layer is {best_hps.get('dense_dropout')},
and the optimal learning rate for the optimizer is {best_hps.get('learning_rate')}.
""")

# Construir o modelo com os melhores hiperparâmetros
model = tuner.hypermodel.build(best_hps)

# Treinar o modelo
history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, callbacks=[stop_early])

# Avaliar o modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
```

```python
# modelo com os melhores parametros RNN acima:
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.optimizers import Adam
from keras.callbacks import Callback
from sklearn.metrics import f1_score, precision_score, recall_score, log_loss, classification_report

# Criar o conjunto de validação
X_train_partial, X_val, y_train_partial, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Definição do callback personalizado para calcular métricas adicionais
class MetricsCallback(Callback):
    def __init__(self, validation_data):
        super().__init__()
        self.validation_data = validation_data

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        y_val, y_true = self.validation_data
        y_pred = np.asarray(self.model.predict(y_val))

        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_true, axis=1)

        logs['val_f1'] = f1_score(y_true_classes, y_pred_classes, average='macro')
        logs['val_precision'] = precision_score(y_true_classes, y_pred_classes, average='macro')
        logs['val_recall'] = recall_score(y_true_classes, y_pred_classes, average='macro')
        logs['val_log_loss'] = log_loss(y_true, y_pred)

        print(f" — val_f1: {logs['val_f1']:.4f} — val_precision: {logs['val_precision']:.4f} — val_recall: {logs['val_recall']:.4f} — val_log_loss: {logs['val_log_loss']:.4f}")

# Inicializando o callback com os dados de validação
metrics_callback = MetricsCallback(validation_data=(X_val, y_val))

# Definindo o modelo RNN com os melhores hiperparâmetros encontrados
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(1, X_train.shape[2])))
model.add(LSTM(96, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(96))
model.add(Dropout(0.1))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_train.shape[1], activation='softmax'))

# Compilando o modelo
optimizer = Adam(learning_rate=0.00042432)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Treinando o modelo com o callback personalizado
history = model.fit(X_train_partial, y_train_partial, epochs=50, batch_size=32, validation_data=(X_val, y_val), callbacks=[metrics_callback])

# Avaliando o modelo no conjunto de teste
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Imprimindo o relatório de classificação e log loss final
print(classification_report(y_true_classes, y_pred_classes))
print(f"Log Loss: {log_loss(y_test, y_pred):.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.optimizers import Adam
from keras.callbacks import Callback, EarlyStopping
from sklearn.metrics import f1_score, precision_score, recall_score, log_loss, classification_report
import tensorflow as tf


# Separar features e rótulos
X = dfmatrix[['LAT_WGS84', 'LONG_WGS84', 'OCC_YEAR', 'MONTH_SIN', 'MONTH_COS', 'HOUR_SIN', 'HOUR_COS', 'OCC_DAY_SIN', 'OCC_DAY_COS', 'OCC_DOY_SIN',  'OCC_DOY_COS', 'OCC_DOW_SIN','OCC_DOW_COS']]
y = dfmatrix[list(onehot_columns)]

# Normalizar as features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Reshape os dados para LSTM (samples, timesteps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Criar o conjunto de validação
X_train_partial, X_val, y_train_partial, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Definição do callback personalizado para calcular métricas adicionais
class MetricsCallback(Callback):
    def __init__(self, validation_data):
        super().__init__()
        self.validation_data = validation_data

    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        y_val, y_true = self.validation_data
        y_pred = np.asarray(self.model.predict(y_val))

        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_true, axis=1)

        logs['val_f1'] = f1_score(y_true_classes, y_pred_classes, average='macro', zero_division=1)
        logs['val_precision'] = precision_score(y_true_classes, y_pred_classes, average='macro', zero_division=1)
        logs['val_recall'] = recall_score(y_true_classes, y_pred_classes, average='macro', zero_division=1)
        logs['val_log_loss'] = log_loss(y_true, y_pred)

        print(f" — val_f1: {logs['val_f1']:.4f} — val_precision: {logs['val_precision']:.4f} — val_recall: {logs['val_recall']:.4f} — val_log_loss: {logs['val_log_loss']:.4f}")

# Inicializando o callback com os dados de validação
metrics_callback = MetricsCallback(validation_data=(X_val, y_val))

# Definindo o Early Stopping com restauração dos melhores pesos
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Definindo o modelo RNN com os melhores hiperparâmetros encontrados
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(1, X_train.shape[2])))
model.add(LSTM(96, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(96))
model.add(Dropout(0.1))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_train.shape[1], activation='softmax'))

# Compilando o modelo
optimizer = Adam(learning_rate=0.00042432)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Treinando o modelo com o callback personalizado e early stopping
history = model.fit(X_train_partial, y_train_partial, epochs=50, batch_size=32, validation_data=(X_val, y_val), callbacks=[metrics_callback, early_stopping_callback])

# Avaliando o modelo no conjunto de teste
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Imprimindo o relatório de classificação e log loss final
print(classification_report(y_true_classes, y_pred_classes, zero_division=1))
print(f"Log Loss: {log_loss(y_test, y_pred):.4f}")
```
