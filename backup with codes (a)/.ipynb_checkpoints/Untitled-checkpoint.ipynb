```python
# Eliminate columns 'X' and 'Y'
crime_data = crime_data.drop(['X', 'Y'], axis=1)

# Set 'OBJECTID' as the new index
crime_data.set_index('OBJECTID', inplace=True)

# Padronização de categorização
crime_data['DIVISION'] = crime_data['DIVISION'].str.lower().str.strip()
crime_data['LOCATION_TYPE'] = crime_data['LOCATION_TYPE'].str.lower()#.str.strip()
#crime_data['PREMISES_TYPE'] = crime_data['PREMISES_TYPE'].str.lower().str.strip()

# Drop rows where OCC was before 2014
crime_data = crime_data[crime_data['OCC_YEAR'] >= 2014]

# Drop rows where the value in the 'DIVISION' column is 'NSA'
crime_data = crime_data[crime_data['DIVISION'] != 'NSA']

# Convert OCC_DATE and REPORT_DATE to datetime
crime_data['OCC_DATE'] = pd.to_datetime(crime_data['OCC_DATE'], errors='coerce')
crime_data['REPORT_DATE'] = pd.to_datetime(crime_data['REPORT_DATE'], errors='coerce')

# Data validation
crime_data = crime_data[crime_data['OCC_DATE'] <= crime_data['REPORT_DATE']]
crime_data = crime_data[(crime_data['REPORT_HOUR'] >= 0) & (crime_data['REPORT_HOUR'] <= 23)]
crime_data = crime_data[(crime_data['OCC_HOUR'] >= 0) & (crime_data['OCC_HOUR'] <= 23)]

# Creating new features
crime_data['REPORT_DAY_OF_WEEK'] = crime_data['REPORT_DATE'].dt.dayofweek
crime_data['OCC_DAY_OF_WEEK'] = crime_data['OCC_DATE'].dt.dayofweek

# Outliers
crime_data = crime_data[(crime_data['LONG_WGS84'] < -78.0000)]
#crime_data = crime_data[(crime_data['LONG_WGS84'] >= -79.639247) & (crime_data['LONG_WGS84'] <= -79.321396)]
#crime_data = crime_data[(crime_data['LAT_WGS84'] >= 43.659776) & (crime_data['LAT_WGS84'] <= 43.853164)]

# Conversão de categorias em números (One-Hot Encoding)
crime_data = pd.get_dummies(crime_data, columns=['DIVISION', 'LOCATION_TYPE', 'PREMISES_TYPE', 'OFFENCE', 'MCI_CATEGORY', 'NEIGHBOURHOOD_158', 'NEIGHBOURHOOD_140'])

# Exibir as primeiras linhas do dataset limpo
crime_data.head()
```
