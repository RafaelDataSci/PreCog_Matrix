## 1. Data Loading...

```python
dfmajorcrimes = pd.read_csv('raw_data/majorcrimes.csv')
#dfbicycle = pd.read_csv('raw_data/bicycle.csv')
dfhomicides = pd.read_csv('raw_data/homicidies.csv')
```

```python
def show_all_columns(df):
    with pd.option_context('display.max_columns', None):
        display(df)
def show_all_rows(df):
    with pd.option_context('display.max_rows', None):
        display(df)
```

```python
import pandas as pd

# Assuming the DataFrames dfhomicides, dfbicycle, dfmajorcrimes, dfshooting_firearm are already defined

# Define the target columns including the ones that need to be added
columns = ['EVENT_UNIQUE_ID','DATASET','OFFENCE', 'MCI_CATEGORY','OCC_HOUR','OCC_DAY','OCC_MONTH','OCC_YEAR','OCC_DOW','OCC_DOY','DIVISION',
           'HOOD_140','NEIGHBOURHOOD_140','HOOD_158','NEIGHBOURHOOD_158','LAT_WGS84','LONG_WGS84']

# Add missing columns with default values
for df in [dfhomicides, dfmajorcrimes]:
    for column in columns:
        if column not in df.columns:
            df[column] = None

# Select columns in the specified order for all DataFrames
#dfbicycle_s = dfbicycle[columns]
dfhomicides_s = dfhomicides[columns]
dfmajorcrimes_s = dfmajorcrimes[columns]

# Set the 'DATASET' column for each DataFrame using .loc to avoid SettingWithCopyWarning
#dfbicycle_s.loc[:, 'DATASET'] = 'BICYCLE'
dfhomicides_s.loc[:, 'DATASET'] = 'HOMICIDES'
dfmajorcrimes_s.loc[:, 'DATASET'] = 'MAJOR_CRIMES'

# Concatenate the DataFrames into a single DataFrame
dfmatrix = pd.concat([dfhomicides_s, dfmajorcrimes_s], ignore_index=True)

# Display the resulting DataFrame
dfmatrix
```

```python
# Agrupar por 'EVENT_UNIQUE_ID' e combinar informações de outras colunas
aggregated_df = dfmatrix.groupby('EVENT_UNIQUE_ID').agg({
    'DATASET': lambda x: list(x),
    'MCI_CATEGORY': lambda x: list(x),
    'OCC_HOUR': lambda x: list(x),
    'OCC_DAY': 'first',
    'OCC_MONTH': 'first',
    'OCC_YEAR': 'first',
    'OCC_DOW': 'first',
    'OCC_DOY': 'first',
    'DIVISION': lambda x: list(x),
    'HOOD_140': lambda x: list(x),
    'NEIGHBOURHOOD_140': lambda x: list(x),
    'HOOD_158': lambda x: list(x),
    'NEIGHBOURHOOD_158': lambda x: list(x),
    'LAT_WGS84': 'first',
    'LONG_WGS84': 'first'
}).reset_index()

# Exibir o DataFrame agregado
aggregated_df
```

```python
def choose_best_record(group):
    # Exemplo: escolher o registro com mais informações (menos valores nulos)
    return group.loc[group.isnull().sum(axis=1).idxmin()]

# Aplicar a função personalizada para cada grupo
cleaned_df = dfmatrix.groupby('EVENT_UNIQUE_ID').apply(choose_best_record).reset_index(drop=True)

# Exibir o DataFrame limpo
cleaned_df
```

```python
# Definindo a ordem de gravidade
severity_order = {
    'Homicide': 1,
    'Robbery': 2,
    'Assault': 3,
    'Auto Theft': 4,
    'Theft Over': 5,
    'Break and Enter': 6
}

# Adicionando uma coluna de gravidade ao DataFrame
dfmatrix['GRAVITY'] = dfmatrix['MCI_CATEGORY'].map(severity_order)

# Função para escolher o registro mais grave
def choose_most_severe(group):
    return group.loc[group['GRAVITY'].idxmin()]

# Aplicar a função personalizada para cada grupo
cleaned_df = dfmatrix.groupby('EVENT_UNIQUE_ID').apply(choose_most_severe).reset_index(drop=True)

# Remover a coluna de gravidade
cleaned_df = cleaned_df.drop(columns=['GRAVITY'])

# Exibir o DataFrame limpo
#import ace_tools as tools; tools.display_dataframe_to_user(name="DataFrame Limpo", dataframe=cleaned_df)

cleaned_df
```
